```python
import pandas as pd
import json
```

```python
with open('/content/T20_Analysis/t20_wc_match_results.json') as f:
    data = json.load(f)
```

```python
data[0]['matchSummary']
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis = 1 , inplace=True)
```

```python
df_match.head()
```

```python
match_dic_ids={}

for index,row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' +  row['team2']
  key2 = row['team2'] + ' Vs ' +  row['team1']

  match_dic_ids[key1] = row['match_id']
  match_dic_ids[key2] = row['match_id']

match_dic_ids
```

### **BATTING SUMMARY**

```python
with open('/content/T20_Analysis/t20_wc_batting_summary.json') as f:
  data = json.load(f)
```

```python
data[0]['battingSummary']
```

```python
all_records=[]
for rec in data:
  all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x : 'out' if len(x) > 0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python

df_batting.head(10)
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply( lambda x : x.replace('†',''))
```

```python
df_batting.head(10)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_dic_ids)
df_batting.head()
```
